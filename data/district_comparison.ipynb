```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("densities.csv", index_col=0)

df_sorted = df.sort_values(by="density", ascending=False)

top10 = df_sorted.head(10)
others_mean = df_sorted.iloc[10:]["density"].mean()

# Add a row for the rest
top10.loc[f"{len(df)-10} other (mean)"] = [None, None, None, others_mean, None, None]

# Plot
plt.figure(figsize=(10, 5))
top10["density"].plot(kind="bar")

plt.title("Top 10 Districts by Pool Density", fontsize=14)
plt.ylabel("Density (pools per km²)")
plt.xlabel("District")
plt.xticks(rotation=45, ha="right")

# Annotate bars
for i, v in enumerate(top10["density"]):
    plt.text(i, v + 4.5, f"{v:.2f}", ha='center', fontsize=9)

plt.tight_layout()
plt.show()
```
